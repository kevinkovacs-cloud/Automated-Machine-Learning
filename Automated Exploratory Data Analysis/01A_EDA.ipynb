<strong>Definición de Percentiles:</strong><br>
El percentil es una medida de posición usada en estadística que indica, una vez ordenados los datos de menor a mayor, el valor de la variable por debajo del cual se encuentra un porcentaje dado de observaciones en un grupo.<br>
Ejemplo: el percentil 20º es el valor bajo el cual se encuentran el 20% de las observaciones y el 80% restante son mayores.
<br><br>
Aparecen citados en la literatura científica por primera vez por Francis Galton en 1885.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;P25 = Q1<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;P50 = Q2 = mediana<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;P75 = Q3<br></p>

<a id = "1.0"></a>
<p style="font-size:14px;color:#16268a;font-family:'Comic Sans MS';text-align:right;border-radius:5px;">
<a href="https://es.wikipedia.org/wiki/Percentil" target=\"_blank\"><i>Referencia Wikipedia - Percentil</i></a></p>
