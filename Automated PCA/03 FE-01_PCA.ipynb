<strong>Principal Component Analysis</strong></p> 

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
El análisis de componentes principales (PCA) es una herramienta que se utiliza para la reducción de dimensiones, pero manteniendo la mayor parte de la información del conjunto de datos original.</p> 
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
PCA puede ser extremadamente útil para ayudar a los algoritmos de aprendizaje automático a realizar buenas predicciones, pero dificulta la posibilidad de hacer inferencias a partir de los datos.</p> 

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Principales usos de PCA:</strong><BR><BR>
<strong>Mejora del rendimiento</strong>: si su algoritmo es demasiado lento debido a las dimensiones, entonces PCA puede acelerarlo.<BR>
<strong>Visualización de datos</strong>: a veces es necesario hacer un diagrama de dispersión en 2 o 3.
</p>     

![00](./img/PCA00.png)

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Como primer paso en el PCA, necesitamos dibujar un nuevo eje que represente la dirección de la varianza máxima (dispersión) de los datos. Esto se denomina <strong>“Primer componente principal”</strong>.</p> 

![00](./img/PCA01.png)

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Se agrega otro eje ortogonal al primer componente principal que representa la siguiente varianza más alta de datos. Esto se denomina <strong>"Segundo componente principal"</strong>.</p>

<strong>Incremental PCA</strong></p>

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px;">
El análisis de componentes principales incrementales (PCA) se utiliza como reemplazo para el análisis de componentes principales (PCA) cuando el conjunto de datos que se va a descomponer es demasiado grande para entrar en memoria.</p>

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px;">
<strong>Parametros</strong>:<BR>
<strong>n_components</strong>, default=None. Number of components to keep.<BR><BR>
<strong>whiten</strong>, default=False. When True (False by default) the components_vectors are divided by n_samples times components_to ensure uncorrelated outputs with unit component-wise variances.<BR><BR>
<strong>copy</strong>, default=True. If False, X will be overwritten. copy=False can be used to save memory but is unsafe for general use.<BR><BR>
<strong>batch_size</strong>, default=None. The number of samples to use for each batch. Only used when calling fit. If batch_size is None, then batch_size is inferred from the data and set to 5 * n_features, to provide a balance between approximation accuracy and memory consumption.</p>

```python
n_components = 2
```

```python
ipca = IncrementalPCA(n_components=n_components, batch_size=10)
```

```python
X_ipca = ipca.fit_transform(X)
```

```python
pca = PCA(n_components=n_components)
```

```python
X_pca = pca.fit_transform(X)
```

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Gráfica de las 2 nuevas características mediante PCA & IPCA</strong></p> 

```python
colors = ['navy', 'turquoise', 'darkorange']
for X_transformed, title in [(X_ipca, "Incremental PCA"), (X_pca, "PCA")]:
    plt.figure(figsize=(8, 8))
    
    for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
        plt.scatter(X_transformed[y == i, 0], X_transformed[y == i, 1], color=color, lw=2, label=target_name)

    if "Incremental" in title:
        err = np.abs(np.abs(X_pca) - np.abs(X_ipca)).mean()
        plt.title(title + " of iris dataset\nMean absolute unsigned error %.6f" % err)
    else:
        plt.title(title + " of iris dataset")
    
    plt.legend(loc="best", shadow=False, scatterpoints=1)
    plt.axis([-4, 4, -1.5, 1.5])
plt.show()    
```

<strong>Kernel PCA</strong></p>

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px;">
<strong>Kernel PCA es capaz de encontrar una proyección de los datos linealmente separables.</strong></p>

```python
X, y = make_circles(n_samples=400, factor=.3, noise=.05)
```

```python
kpca = KernelPCA(kernel="rbf", fit_inverse_transform=True, gamma=10)
```

```python
X_kpca = kpca.fit_transform(X)
```

```python
X_back = kpca.inverse_transform(X_kpca)
```

```python
pca = PCA()
X_pca = pca.fit_transform(X)
```

```python
plt.figure()
plt.subplot(2, 2, 1, aspect='equal')
plt.title("Original space")
reds = y == 0
blues = y == 1

plt.scatter(X[reds, 0], X[reds, 1], c="red", s=20, edgecolor='k')
plt.scatter(X[blues, 0], X[blues, 1], c="blue", s=20, edgecolor='k')
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")

X1, X2 = np.meshgrid(np.linspace(-1.5, 1.5, 50), np.linspace(-1.5, 1.5, 50))

X_grid = np.array([np.ravel(X1), np.ravel(X2)]).T

Z_grid = kpca.transform(X_grid)[:, 0].reshape(X1.shape)

plt.contour(X1, X2, Z_grid, colors='grey', linewidths=1, origin='lower')

plt.subplot(2, 2, 2, aspect='equal')
plt.scatter(X_pca[reds, 0], X_pca[reds, 1], c="red", s=20, edgecolor='k')
plt.scatter(X_pca[blues, 0], X_pca[blues, 1], c="blue", s=20, edgecolor='k')
plt.title("Projection by PCA")
plt.xlabel("1st principal component")
plt.ylabel("2nd component")

plt.subplot(2, 2, 3, aspect='equal')
plt.scatter(X_kpca[reds, 0], X_kpca[reds, 1], c="red", s=20, edgecolor='k')
plt.scatter(X_kpca[blues, 0], X_kpca[blues, 1], c="blue", s=20, edgecolor='k')
plt.title("Projection by KPCA")
plt.xlabel("1st principal component in space induced by $\phi$")
plt.ylabel("2nd component")

plt.subplot(2, 2, 4, aspect='equal')
plt.scatter(X_back[reds, 0], X_back[reds, 1], c="red", s=20, edgecolor='k')
plt.scatter(X_back[blues, 0], X_back[blues, 1], c="blue", s=20, edgecolor='k')
plt.title("Original space after inverse transform")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")

plt.subplots_adjust(0.02, 0.10, 0.98, 0.94, 0.04, 0.35)
plt.show()
```
