<strong>El Análisis de Componentes Principales (PCA)</strong> identifica la combinación de atributos que representan la mayor varianza entre los datos.</p>
<p style="font-size:17px;color:#16268a;font-family:'Comic Sans MS';text-align:left;border-radius:5px;">
<strong>El Análisis Discriminante Lineal (LDA)</strong> intenta identificar los atributos que representan la mayor varianza entre las clases.</p>

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
pca = PCA(n_components=3)
```

```python
X_r = pca.fit(X).transform(X)
```

```python
lda = LinearDiscriminantAnalysis(n_components=3)
```

```python
X_r2 = lda.fit(X, y).transform(X)
```

```python
print('explained variance ratio (first two components): %s' % str(pca.explained_variance_ratio_))
```

```python
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2
```

```python
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw, label=target_name)
```

```python
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA IRIS dataset')
plt.figure()
```

```python
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color, label=target_name)
```

```python
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA IRIS dataset')
```

```python
plt.show()
```
